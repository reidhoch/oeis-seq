```python
%matplotlib inline

from itertools import count, islice
from typing import Callable, Iterable

from pandas import DataFrame
from seaborn import lineplot, set_theme

from random import choice
from oeis.registry import registry

set_theme(style="darkgrid")

name, seq = choice(list(registry.sequences.items()))

df: DataFrame = DataFrame()
df["index"] = list(islice(count(), 10))
df["values"] = list(islice(seq(), 10))

lineplot(x="index", y="values", label=name, data=df)
```
